## Cleaning

```python
df.dropna(axis=0, how='any', subset=['Order Lines/Subtotal'], inplace=True)

def remove_apostrophe(text):
    if type(text) is not float:
        return text.replace("'", "")
    return text

float_cols = ['Order Lines/Subtotal', 'Total', 'Order Lines/Unit Price']
int_cols = ['Order Lines/Quantity', 'Loyalty Points']
df[float_cols+int_cols] = df[float_cols+int_cols].applymap(remove_apostrophe).astype('float64')


df.fillna(method='ffill', inplace=True)
df[int_cols] = df[int_cols].astype('int32')
# df['Loyalty Points'] = df['Loyalty Points'].astype('int32')
df.head()
```

```python
df.isnull().sum().sum()
```

## Finding total discount

```python
# Finding Discount %
df['TotalDiscount'] = df['Order Lines/Unit Price'] * df['Order Lines/Quantity'] * (df['Order Lines/Discount (%)']/100)
# Fixed Discount Column 
df['TotalDiscount'] += df['Order Lines/Discount Fixed']

#Discount found as a product
def isdiscount(text):
    if 'Discount' in text:
        return True
    else: return False
    
discount_items = df[df.loc[:,'Order Lines/Product/Name'].map(isdiscount)][['Order Lines/Subtotal']]*-1
df.loc[discount_items.index, ['TotalDiscount']] += discount_items.values

df['TotalDiscount'].sum()
```

### Payment Methods

```python
df['Payments/Journal/Journal Name'].unique()
```

```python
def simple_payment_method(text):
    types = ['Credit - Talabat', 'Credit - Instashop', 'Card', 'Cash',
             'Credit Note', 'Credit- HOMIEZ','Credit:-Feehla']
    for t in types:
        if t in text:
            return t
        
  
df['Payments/Journal/Journal Name'] = df['Payments/Journal/Journal Name'].map(simple_payment_method)
df['Payments/Journal/Journal Name'].unique()
```

### Type of Sale

Grooming Vs. Product Vs. App Sale

```python
def type_of_sale(row):
    
    if row['Order Lines/Product/Product Category'] == 'Grooming Service':
        return 'Grooming'

    app_types = ['Talabat', 'Insta', 'HOMIEZ']
    for typ in app_types:
        if typ in row['Payments/Journal/Journal Name']:
            return typ    
    return 'Product'

df['SaleType'] = df.apply(type_of_sale, axis=1)
df['SaleType'].value_counts()
```

## Separating The Type of Orders


App Vs. Product Vs. Grooming

```python
# Separate App Order from the rest
app_order_indices = order_totals['SaleType'].map(lambda typ: typ not in ['Product', 'Grooming'])
app_orders = order_totals[app_order_indices].copy()



# Shop Orders - > Product + Grooming
shop_orders = order_totals[~app_order_indices].copy()
product_orders = shop_orders[shop_orders['SaleType'] == 'Product'].copy()
groom_orders = shop_orders[shop_orders['SaleType'] == 'Grooming'].copy()

product_orders.drop('SaleType', inplace=True, axis=1)
groom_orders.drop('SaleType', inplace=True, axis=1)
app_orders.drop('SaleType', inplace=True, axis=1)

product_orders.shape[0] + app_orders.shape[0] + groom_orders.shape[0]
```

### Get Pivot Tables

A really handy function

```python
def get_pivot_table(df, columns, values, indices):
    agg_cols = ['Loyalty Points', 'TotalDiscount', 'Total']
    agg_part = df[indices+agg_cols].groupby(indices).sum().reset_index()
    pivot_table = pd.pivot(df, index=indices, columns=columns,
                           values=values).reset_index()
    # Merge Them
    return pd.merge(pivot_table,agg_part,how='inner',on=indices)
```

### Grooming & Products Section

```python
indices = ['Branch','Date','Salesperson']

products_table = get_pivot_table(product_orders, ['PaymentType'], 'Total', indices)
groom_table = get_pivot_table(groom_orders, ['PaymentType'], 'Total', indices)
```

## Combining the Tables

```python
merg1 = pd.merge(products_table, groom_table, how='outer', on=indices, suffixes=('_P', '_G'))
whole_df = pd.merge(merg1, app_table, how='outer', on=indices, suffixes=('_S', '_App'))
whole_df.fillna(0., inplace=True)
whole_df.info()
whole_df.head()
```

### Get Rid of Salesperson Column

```python
quarter_summ = whole_df.groupby(['Branch', 'Date']).sum().reset_index()
quarter_summ
```

## Mapping Columns to Excel Template

```python
excel_sheet_map = { # (Row, Col) Index
    (4,1): ['Date'],
    (4,4): ['Cash_P', 'Card_P', 'AMEX', 'Loyalty Points_P', 'TotalDiscount_P', 'Return_P'], #Product Sales
    (4,12): ['Cash_G', 'Card_G', 'AMEX', 'Loyalty Points_G', 'TotalDiscount_G', 'Return_G'], # Grooming Sales
    (4,26): ['Credit - Talabat', 'Credit- HOMIEZ', 'Baqalaat', 'IS - Cash','Credit - Instashop', 'Total']
}
# Create columns if they don't exist
for cols in excel_sheet_map.values():
    for c in cols:
        if c not in quarter_summ.columns:
            print(c)
            quarter_summ[c] = 0.0

            
```

## Write to Excelbook

```python
from openpyxl import Workbook, load_workbook

from openpyxl.utils.dataframe import dataframe_to_rows

wb = load_workbook(filename = 'Branch Daily Sales Report Sample.xlsx')
ws = wb['JUN']

def insert_xltable(ws, df, df_to_xlmap):
    # Create Columns if they don't exist
    for (start_row, start_col), cols in df_to_xlmap.items():
        subset = df[cols].values
        
        width = subset.shape[1]
        height = subset.shape[0]
        
        for row in range(start_row, start_row+height):
            for col in range(start_col, start_col+width):
                ws.cell(row=row, column=col,
                        value=subset[row-start_row, col-start_col])
        

# Isolate the right month
june_orders = quarter_summ[(quarter_summ['Date'] >= '2021-06-01') & (quarter_summ['Date'] <= '2021-06-30')]
# Get the right branch
manama_june_orders = june_orders[june_orders['Branch'] == 'Manama']


insert_xltable(ws, manama_june_orders, excel_sheet_map)
wb.save('processed.xlsx')
```
