## Cleaning

```python
df.dropna(axis=0, how='any', subset=['Order Lines/Subtotal'], inplace=True)

def remove_apostrophe(text):
    if type(text) is not float:
        return text.replace("'", "")
    return text

float_cols = ['Order Lines/Subtotal', 'Total', 'Order Lines/Unit Price']
int_cols = ['Order Lines/Quantity', 'Loyalty Points']
df[float_cols+int_cols] = df[float_cols+int_cols].applymap(remove_apostrophe).astype('float64')


df.fillna(method='ffill', inplace=True)
df[int_cols] = df[int_cols].astype('int32')
# df['Loyalty Points'] = df['Loyalty Points'].astype('int32')
df.head()
```

```python
df.isnull().sum().sum()
```

## Finding total discount

```python
# Finding Discount %
df['TotalDiscount'] = df['Order Lines/Unit Price'] * df['Order Lines/Quantity'] * (df['Order Lines/Discount (%)']/100)
# Fixed Discount Column 
df['TotalDiscount'] += df['Order Lines/Discount Fixed']

#Discount found as a product
def isdiscount(text):
    if 'Discount' in text:
        return True
    else: return False
    
discount_items = df[df.loc[:,'Order Lines/Product/Name'].map(isdiscount)][['Order Lines/Subtotal']]*-1
df.loc[discount_items.index, ['TotalDiscount']] += discount_items.values

df['TotalDiscount'].sum()
```

```python
def simple_payment_method(text):
    types = ['Talabat', 'Insta', 'Card', 'Cash',
             'Credit Note', 'Credit- HOMIEZ','Feehla']
    for t in types:
        if t in text:
            return t
        
  
df['Payments/Journal/Journal Name'] = df['Payments/Journal/Journal Name'].map(simple_payment_method)
df['Payments/Journal/Journal Name'].unique()
```

### Grooming Column

```python
def get_pivot_table(df, columns, values, indices):
    grp = df[indices + [columns, values]].groupby(indices + [columns]).sum()
    grp = grp.reset_index()

    piv_table = pd.pivot(grp, index=indices, columns = columns,values=values)
    return piv_table.reset_index()

def get_sumgroupby(df, agg_cols, indices):
    
    return df[indices+agg_cols].groupby(indices).sum().reset_index()
```
