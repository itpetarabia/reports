## Cleaning

```python
df.dropna(axis=0, how='any', subset=['Order Lines/Subtotal'], inplace=True)

def remove_apostrophe(text):
    if type(text) is not float:
        return text.replace("'", "")
    return text

float_cols = ['Order Lines/Subtotal', 'Total', 'Order Lines/Unit Price']
int_cols = ['Order Lines/Quantity']
df[float_cols+int_cols] = df[float_cols+int_cols].applymap(remove_apostrophe).astype('float64')

# Ensure that NA values in Order Subtotal Field doesn't mess with the ffill method
numbers_cols = ["Order Lines/Unit Price","Order Lines/Quantity","Order Lines/Discount (%)","Order Lines/Discount Fixed","Order Lines/Subtotal"]
df[numbers_cols] = df[numbers_cols].fillna(0.)

df.fillna(method='ffill', inplace=True)
df[int_cols] = df[int_cols].astype('int32')

df.head(20)
```

## Finding Total Loyalty Points & Discounts

```python
# df['Loyalty Points'] = 0


def get_loyalty_and_discounts(row):
    loyalty,discount = 0,0
    if row['Order Lines/Unit Price'] < 0:
        if ('Loyalty' in row['Order Lines/Product/Name']) & ('Discount' in row['Order Lines/Product/Product Category']):
            loyalty = 5 # Fixed value of 5BD
        else:
            discount = -1.0 * row['Order Lines/Unit Price'] * row['Order Lines/Quantity']
    return (loyalty,discount)
    
df[['Loyalty Points','TotalDiscount']] = df.apply(get_loyalty_and_discounts, result_type='expand',axis=1)
# df.head()
```

## Finding *real* total discount

```python
# Finding Discount %
df['TotalDiscount'] += df['Order Lines/Unit Price'] * df['Order Lines/Quantity'] * (df['Order Lines/Discount (%)']/100)
# Fixed Discount Column 
df['TotalDiscount'] += df['Order Lines/Discount Fixed']

df['TotalDiscount'].sum()
```

## Finding Total Refund

```python

def get_refunds(row):
    qty = row['Order Lines/Quantity']
    if qty < 0:
        return -1*qty * row['Order Lines/Unit Price']
    else:
        return 0.

    
df['Refunds'] = df.apply(get_refunds, axis=1)
df['Refunds'].sum()
```

### Payment Methods

```python
df['Payments/Journal/Journal Name'].unique()
```

```python
def simple_payment_method(text):
    types = ['Credit - Talabat', 'Credit - Insta', 'Card', 'Cash',
             'Credit Note', 'Credit- HOMIEZ','Credit:-Feehla']
    for t in types:
        if t in text:
            return t
        
  
df['Payments/Journal/Journal Name'] = df['Payments/Journal/Journal Name'].map(simple_payment_method)
df['Payments/Journal/Journal Name'].unique()
```

### Type of Sale

Grooming Vs. Product Vs. App Sale

```python
def type_of_sale(row):
    
    if row['Order Lines/Product/Product Category'] == 'Grooming Service':
        return 'Grooming'

    app_types = ['Talabat', 'Insta', 'HOMIEZ']
    for typ in app_types:
        if typ in row['Payments/Journal/Journal Name']:
            return typ    
    return 'Product'

df['SaleType'] = df.apply(type_of_sale, axis=1)
df['SaleType'].value_counts()
```

### Get Pivot Tables

A really handy function

```python
def get_pivot_table(df, columns, values, indices):
    agg_cols = ['Loyalty Points', 'TotalDiscount', 'Total']
    agg_part = df[indices+agg_cols].groupby(indices).sum().reset_index()
    pivot_table = pd.pivot(df, index=indices, columns=columns,
                           values=values).reset_index()
    # Merge Them
    return pd.merge(pivot_table,agg_part,how='inner',on=indices)
```

### Grooming & Products Section

```python
indices = ['Branch','Date','Salesperson']

products_table = get_pivot_table(product_orders, ['PaymentType'], 'Total', indices)
groom_table = get_pivot_table(groom_orders, ['PaymentType'], 'Total', indices)
```
